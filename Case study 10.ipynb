# Data Cleaning

```python
# drop "veil-type", 'veil-color','gill-attachment','stalk-root',features were reduced to 19 columns
data = data.drop(['veil-color','gill-attachment','stalk-root'], axis = 1)
data.describe()
```

```python
X = data.iloc[:,1:]
y = data.iloc[:,0].values
X
```

```python
y
```

# Encode categorical data

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
for col in data.columns:
    data[col] = labelencoder.fit_transform(data[col])
 
```

```python
data
```

# Splitting the data into train and test

```python
#splitting training data in test, train, test size is 20% of whole train data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

# Logistic Regression

```python
#importing Logistic Regression 
from sklearn.linear_model import LogisticRegression
#creating our model
model1=LogisticRegression()
#applying Logistic Regression on train data
model1.fit(X_train, y_train)
```

```python
prediction1= model1.predict(X_test)
prediction1
```

# Accuracy score

```python
from sklearn.metrics import accuracy_score 
accuracy_score(prediction1,y_test)
```

# Evaluation using confusion matrix

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, prediction1 )
```

Note: The confusion matrix here describes how many correct predictions and wrong predictions have been made
for class=0 and class=1. Total wrong predictions=0+0=0

# Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
model_naive = GaussianNB()
model_naive.fit(X_train, y_train)
```

```python
y_predict=model_naive.predict(X_test)
y_predict
```

# Evaluation using confusion matrix and accuracy score

```python
from sklearn.metrics import confusion_matrix,accuracy_score 
```

```python
confusion_matrix(y_test, y_predict )
```

Note: The confusion matrix here describes how many correct predictions and wrong predictions have been made
for class=0 and class=1. Total wrong predictions=0+24=24

```python
accuracy_score(y_test,y_predict)
```
